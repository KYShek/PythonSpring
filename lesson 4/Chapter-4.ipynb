```python
import matplotlib.pyplot as plt
from pandas import DataFrame,Series
```

```python
#4.1.2.1
income=[1.47,1.62,1.78,1.94,2.38,2.60]
data=DataFrame({'Income':income},index=['2012','2013','2014','2015','2016','2017'])
data.plot(title='Income chart',linewidth=2,marker='s',linestyle='--',color='k',grid=True,use_index=True,ylim=[0,3],xlim=[0,5])
plt.annotate('Largest!',xy=(5,2.6),xytext=(3.25,2.55),arrowprops=dict(arrowstyle='->'))
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income(RMB Ten Thousand)',fontsize=12)
plt.savefig('2012-2017income.jpg',dpi=400,bbox_inches='tight')
plt.show()
```

```python
#4.1.2.2
data2=Series(income,index=['2012','2013','2014','2015','2016','2017'])
fig=plt.figure(figsize=(6,6))
plt.subplots_adjust(wspace=0.4,hspace=0.4)
ax1=fig.add_subplot(2,2,1)
plt.title('line chart',fontsize=10)
plt.xlabel('Year')
plt.ylabel('Income')
ax1.plot(data2)
ax2=fig.add_subplot(2,2,2)
data.plot(kind='box',fontsize='small',xticks=[],ax=ax2)
plt.title('box-whisker plot',fontsize=10)
plt.xlabel('2012-2017')
plt.ylabel('Income')
ax3=fig.add_subplot(2,1,2)
data.plot(kind='bar',use_index=True,fontsize='small',ax=ax3)
plt.title('bar chart',fontsize=10)
plt.xlabel('Year')
plt.ylabel('Income')
```

```python
#4.2.2.1
import pandas as pd
data=pd.read_csv('data/High-speed rail.csv',index_col='Country')
data['Operation'].plot(kind='bar',title='Operation Mileage',rot=60)
plt.ylabel('Mileage(km)')
plt.annotate('Longest!',xy=(0,20000),xytext=(0.8,19800),arrowprops=dict(arrowstyle='-'))
```
