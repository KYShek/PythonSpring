```python
#1
import pandas as pd
data=pd.read_csv('data/wine.data',header=None)
data
x=data.loc[:,1:13].values.astype(float)
y=data.loc[:,0].values.astype(int)
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.5,random_state=1)
from sklearn import tree
clf=tree.DecisionTreeClassifier()
clf=clf.fit(x,y)
clf.score(x,y)
```

```python
clf.score(x_test,y_test)
predicted_y=clf.predict(x_test)
from sklearn import metrics
metrics.classification_report(y_test,predicted_y)
metrics.confusion_matrix(y_test,predicted_y)
```

```python
#2
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

data=pd.read_excel('data/housePrice.xlsx',index_col=0)
data

```

```python

data.plot(kind='scatter',x='unit size',y='price',title='price and unit size')
plt.xlabel("unit size")
plt.ylabel('price')
plt.show()
data.plot(kind='scatter',x='room',y='price',title='price and room')
plt.xlabel("room")
plt.ylabel('price')
plt.show()
data.plot(kind='scatter',x='year',y='price',title='price and year')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
data.plot(kind='scatter',x='living room',y='price',title='price and living room')
plt.xlabel('living room')
plt.ylabel('price')
plt.show()
```

```python
X=data.iloc[:,[0,1]].values.astype(float)
y=data['price'].values.astype(int)

from sklearn import model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.30,random_state=1)

from sklearn.linear_model import LinearRegression
linregTr= LinearRegression()
linregTr.fit(X_train,y_train)
print(linregTr.intercept_,linregTr.coef_)

from sklearn import metrics
y_test_pred=linregTr.predict(X_test)
print(y_test_pred)

test_err=metrics.mean_squared_error(y_test,y_test_pred) 
print('The mean squar error of test is: {:.2f}'.format(test_err))
decision_score=linregTr.score(X_test,y_test) 
print('The decision coeficient is: {:.2f}'.format(decision_score))
```
